```python
import pandas as pd
```

```python
data = pd.read_csv("/kaggle/input/home-loan-status-prediction/data.csv")
```

```python
#to see the rows of the data frame
data.head()
```

```python
#to get information about the dataset
data.info()
```

```python
#to get statitstical information about the dataset
data.describe
```

```python
#to check if there any duplicates in the data
data.duplicated().sum()
```

```python
#inorder to see duplicates
duplicate_rows = data[data.duplicated()]
duplicate_rows
```

```python
#to drop duplicates
data = data.drop_duplicates(ignore_index=True)
data
```

```python
#to find the missing values
data.isnull().sum().sort_values(ascending=False)

# isnull() - To find the null values
# sum() - To find the sum of null values for each attribute
# sort_values() - To sort the values
```

```python
#the percentage of values missing
(data.isnull().sum().sum()/data.size)*100
```

```python
#dropping rows with null values
data_dropna = data.dropna()
data_dropna
```

```python
#dropping column with null values
data_dropcol = data.dropna(axis=1)
data_dropcol
```

```python
#bfill can be used to fill null values with values present in the next row of the dataframe
data_bfilled = data.bfill()
data_bfilled.isnull().sum()
```

```python
#ffill can be used to fill null values with values present in the previous row of the dataframe
data_ffilled = data.ffill()
data_ffilled.isnull().sum()
```

```python
#using fillna method to fill the Na or Nan values of a dataframe
data_fillna = data.fillna(0)
data_fillna.info()
```
